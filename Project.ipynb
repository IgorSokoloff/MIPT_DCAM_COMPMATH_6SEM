### Task:

Create splitting scheme for the two-dimensional heat equation:
$$
\frac{\partial u}{\partial t} = \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2},\qquad 0\leqslant t\leqslant 0.5,\qquad 0\leqslant x\leqslant 1,\qquad 0\leqslant y\leqslant 1$$

$$
u\big|_{t=0} = \begin{cases}
\frac{1}{h^2}, x=0.5,y=0.5 \\ 
0, x\neq 0.5 ,y\neq 0.5\\ 
\end{cases}
$$

$$u\big|_{x=0} = 0$$
$$u\big|_{x=1} = 0$$
$$u\big|_{y=0} = 0$$
$$u\big|_{y=1} = 0$$

using next scheme:
$$\frac{u_{m,n}^{p+^1/_2}-u_{m,n}^{p}}{\tau} = \varkappa(\Lambda_{xx}u_{m,n}^{p}+\Lambda_{yy}u_{m,n}^{p+^1/_2})$$
$$\frac{u_{m,n}^{p+1}-u_{m,n}^{p+^1/_2}}{\tau} = \varkappa(\Lambda_{xx}u_{m,n}^{p+1}+\Lambda_{yy}u_{m,n}^{p+^1/_2})$$
$$u_{0,n}^{p}=0$$
$$u_{m,0}^{p}=0$$

$$
u_{m,n}^{0} = \begin{cases}
\frac{1}{h^2}, x=0.5,y=0.5 \\ 
0, x\neq 0.5 ,y\neq 0.5\\ 
\end{cases}
$$

where
$$\Lambda_{xx}u_{m,n}^{p} = \frac{u_{m+1,n}^{p}-2u_{m,n}^{p}+u_{m-1,n}^{p}}{h^2}$$
$$\Lambda_{yy}u_{m,n}^{p} = \frac{u_{m,n+1}^{p}-2u_{m,n}^{p}+u_{m,n-1}^{p}}{h^2}$$

$$\varkappa = \frac{1}{2}$$ 
$p = 0,1, ...,P-1$

Let $M = 100,\;h = \frac{1}{M},\;\tau=\frac{h}{2},\; \sigma=\frac{\tau}{h}=\frac{1}{2}$


$P\tau=T$

### Solve:

$$(E - \tau\varkappa\Lambda_{yy})u_{m,n}^{p+^1/_2} = (E + \tau\varkappa\Lambda_{xx})u_{m,n}^{p}$$
$$(E - \tau\varkappa\Lambda_{xx})u_{m,n}^{p+^1} = (E + \tau\varkappa\Lambda_{yy})u_{m,n}^{p+^1/_2}$$

Let $\sigma = \tau\varkappa/h^2$


Then
$$u_{m,n}^{p+^1/_2} - \sigma(u_{m,n-1}^{p+^1/_2}-2u_{m,n}^{p+^1/_2}+u_{m,n+1}^{p+^1/_2}) = u_{m,n}^{p} + \sigma(u_{m+1,n}^{p}-2u_{m,n}^{p}+u_{m-1,n}^{p})$$
$$u_{m,n}^{p+1} - \sigma(u_{m-1,n}^{p+1}-2u_{m,n}^{p+1}+u_{m+1,n}^{p+1}) = u_{m,n}^{p+^1/_2} + \sigma(u_{m,n-1}^{p^1/_2}-2u_{m,n}^{p^1/_2}+u_{m,n+1}^{p^1/_2})$$


$$(-\sigma)u_{m,n-1}^{p+^1/_2}+(1+2\sigma)u_{m,n}^{p+^1/_2} + (-\sigma)u_{m,n+1}^{p+^1/_2} = (\sigma)u_{m-1,n}^{p}+(1-2\sigma)u_{m,n}^{p} + (\sigma)u_{m+1,n}^{p}$$
$$(-\sigma)u_{m-1,n}^{p+1}+(1+2\sigma)u_{m,n}^{p+1} + (-\sigma)u_{m+1,n}^{p+1} = (\sigma)u_{m,n-1}^{p+^1/_2}+(1-2\sigma)u_{m,n}^{p+^1/_2} + (\sigma)u_{m,n+1}^{p+^1/_2}$$

Using boundary conditions we have

$$(1+2\sigma)u_{m,1}^{p+^1/_2} + (-\sigma)u_{m,2}^{p+^1/_2} = (\sigma)u_{m-1,1}^{p}+(1-2\sigma)u_{m,1}^{p} + (\sigma)u_{m+1,1}^{p}$$
$$(1+2\sigma)u_{1,n}^{p+1} + (-\sigma)u_{2,n}^{p+1} = (\sigma)u_{1,n-1}^{p+^1/_2}+(1-2\sigma)u_{1,n}^{p+^1/_2} + (\sigma)u_{1,n+1}^{p+^1/_2}$$

So we have systems of linear equantions:


$
\qquad
\begin{pmatrix}
1+2\sigma & -\sigma & 0 & \dots & 0 \\
-\sigma & 1+2\sigma & -\sigma & \dots & 0\\
\ldots &\ldots &\ldots &\ddots &\ldots &\\
 0 & \dots & -\sigma & 1+2\sigma & -\sigma \\
0 &\dots & 0 &-\sigma & 1+2\sigma \\
\end{pmatrix}
$
$
\begin{pmatrix}
u_{m,1}^{p+^1/_2}\\
u_{m,2}^{p+^1/_2}\\
\vdots\\
u_{m,M-2}^{p+^1/_2}
\end{pmatrix}
$
=
$
\begin{pmatrix}
(\sigma)u_{0,n}^{p}+(1-2\sigma)u_{1,n}^{p}+(\sigma)u_{2,n}^{p}\\
(\sigma)u_{1,n}^{p}+(1-2\sigma)u_{2,n}^{p}+(\sigma)u_{3,n}^{p}\\
\vdots\\
(\sigma)u_{M-3,n}^{p}+(1-2\sigma)u_{M-2,n}^{p}+(\sigma)u_{M-1,n}^{p}\\
\end{pmatrix}
$


$
\qquad
\begin{pmatrix}
1+2\sigma & -\sigma & 0 & \dots & 0 \\
-\sigma & 1+2\sigma & -\sigma & \dots & 0\\
\ldots &\ldots &\ldots &\ddots &\ldots &\\
 0 & \dots & -\sigma & 1+2\sigma & -\sigma \\
0 &\dots & 0 &-\sigma & 1+2\sigma \\
\end{pmatrix}
$
$
\begin{pmatrix}
u_{1,n}^{p+1}\\
u_{2,n}^{p+1}\\
\vdots\\
u_{M-2,n}^{p+1}
\end{pmatrix}
$
=
$
\begin{pmatrix}
(\sigma)u_{m,0}^{p+^1/_2}+(1-2\sigma)u_{m,1}^{p+^1/_2}+(\sigma)u_{m,2}^{p+^1/_2}\\
(\sigma)u_{m,1}^{p+^1/_2}+(1-2\sigma)u_{m,2}^{p+^1/_2}+(\sigma)u_{m,3}^{p+^1/_2}\\
\vdots\\
(\sigma)u_{m,M-3}^{p+^1/_2}+(1-2\sigma)u_{m,M-2}^{p+^1/_2}+(\sigma)u_{m,M-1}^{p+^1/_2}\\
\end{pmatrix}
$

which we can solve using tridiagonal matrix algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
%matplotlib inline
```

```python
## parameters

kappa = 0.5
M = 100
h = 1 / M  # for x and y axes
tau = h**2 / 2
sigma = tau*kappa / h**2
T = 0.5
Xmax = Ymax = 1

number_of_plots = 10

def plot2(u_plot, times, shape):
    colors = [(0, 0, 1), (1, 0, 0)]  # R -> G -> B
    n_bins = np.full((10), 1000)  # Discretizes the interpolation into bins
    cmap_name = 'my_list'
    fig, axs = plt.subplots(shape[0], shape[1], figsize=(20, 20))
    #fig.subplots_adjust(left=0.02, bottom=0.06, right=0.95, top=0.94, wspace=0.05, hspace=0.05)
    fig.subplots_adjust(left=0.02, bottom=0.06, right=0.95, top=0.94, wspace=0.05)
    #fig.subplots_adjust(wspace=0.5, hspace=0.5)
    i = 0
    for n_bin, ax in zip(n_bins, axs.ravel()):
        # Create the colormap
        cm = LinearSegmentedColormap.from_list(cmap_name, colors, N=n_bin)

        # Fewer bins will result in "coarser" colomap interpolation
        im = ax.imshow(u_plot[i], interpolation='nearest', origin='lower', cmap=cm)

        ax.set_title("t = "+str(times[i]))

        fig.colorbar(im, ax=ax)
        i += 1

    cm = LinearSegmentedColormap.from_list(cmap_name, colors, N=100)

    # Fewer bins will result in "coarser" colomap interpolation
    #im = ax.imshow(u, interpolation='nearest', origin='lower', cmap=cm)
    plt.show()

def plot(u):
    fig = plt.figure(figsize=(20, 20))
    ax = fig.add_subplot(111)

    im = ax.matshow(u)
    fig.colorbar(im, ax=ax)
    plt.show()


def delta(x, y):  #   -  
    return 1 / h ** 2 if (x == Xmax / 2 and y == Ymax / 2) else 0

    sig2 = 0.05**2;
    r2 = (x-Xmax/2)**2 + (y-Ymax/2)**2


    return np.exp(-r2 / sig2 / 2) / np.sqrt(2*np.pi*sig2)
    #return 1 / h ** 2 if (x == 0.02 and y == 0.02) else 0

def boundary(t):  #   
    return 0.0

def solve_tdm(a, b, c, f):
    n = len(a)
    P = np.zeros(n); Q = np.zeros(n); x = np.zeros(n)
    P[0] = -c[0] / b[0]
    Q[0] = f[0] / b[0]
    for k in range(n-1):
        P[k+1] = -c[k+1] / (a[k+1] * P[k] + b[k+1])
        Q[k+1] = (f[k+1] - a[k+1] * Q[k]) / (a[k+1] * P[k] + b[k+1])
    x[n-1] = Q[n-1]
    for k in reversed(range(n-1)):
        x[k] = P[k] * x[k+1] + Q[k]
    return x


def scheme(u, sigma, M):
    u_new = np.copy(u)

    a = np.zeros(M-2)
    b = np.zeros(M-2)
    c = np.zeros(M-2)
    f1 = np.zeros(M-2)
    f2 = np.zeros(M-2)
    a[0] = 0
    a[1:] = -sigma
    b[:] = 1 + 2 * sigma
    c[-1] = 0
    c[:-1] = -sigma

    # 
    for j in range(1, M - 1):
        for i in range(1, M - 1):
            f1[i - 1] = sigma * u[i][j - 1] + (1 - 2*sigma)*u[i][j] + sigma*u[i][j + 1]

        output1 = solve_tdm(a, b, c, f1)
        #output2 = solve_banded((1, 1), G, f1)
        u_new[1:-1, j] = output1

    u = u_new.copy()
    # 
    for i in range(1, M - 1):
        for j in range(1, M - 1):
            f2[j - 1] = sigma * u[i-1][j] + (1 - 2*sigma)*u[i][j] + sigma*u[i+1][j]

        output1 = solve_tdm(a, b, c, f2)
        #output2 = solve_banded((1, 1), G, f2)
        u_new[i, 1:-1] = output1

  #  plot2(u_new)
    return u_new


def solve_heat_conduction(u0, g, method, kappa=0.5, M=100, h=1/M, tau=h/2, tmax=0.5):

    X = Y = np.arange(0, Xmax, h)

    u = np.zeros((len(Y), len(X)))

    for i in range (len(Y)):
        for j in range(len(X)):
            u[i][j] = delta(X[j], Y[i])

    t = 0.0
    steps = 0.0

    #plot2(u)

    u_plot = np.zeros((number_of_plots, len(Y), len(X)))
    #u_plot[0] = u.copy()

    i = 1
    j = 0
    times = np.zeros(number_of_plots)

    while t < tmax:
        if t + tau > tmax: tau = tmax - t + 1e-14
        unext = np.empty_like(u)
        #  
        unext[0] = boundary(t + tau)
        unext[M - 1] = boundary(t + tau)
        unext[:, 0] = boundary(t + tau)
        unext[:, M - 1] = boundary(t + tau)

        #       
        unext = method(u, sigma, M)

        u = unext.copy()  #      

        if (i % 10 == 0):
            u_plot[j] = u.copy()
            times[j] = t
            j += 1

        t += tau;
        steps += 1
        i += 1


    #plot(u)
    return X, Y, u_plot, times


X, Y, U_plot, Times= solve_heat_conduction(delta, boundary,scheme)

plot2(U_plot, Times, (5, 2))
#plot(U)
```
